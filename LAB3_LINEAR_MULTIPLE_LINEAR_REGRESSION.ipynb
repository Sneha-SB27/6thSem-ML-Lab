```python
import numpy as np

# Data
X = np.array([1,2,3,4]).reshape(-1,1)
Y = np.array([2,4,5,9]).reshape(-1,1)

# Add intercept
X_design = np.hstack([np.ones((X.shape[0],1)), X])

# Normal equation
beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y
print("Intercept:", beta[0][0])
print("Slope:", beta[1][0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('canada_per_capita_income.csv')

X = df[['year']].values
Y = df.iloc[:,1].values.reshape(-1,1)

X_design = np.hstack([np.ones((X.shape[0],1)), X])

beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

# Prediction
pred = np.array([1,2020]) @ beta
print("Predicted income:", pred[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('salary.csv')

# Remove missing rows if any
df = df.dropna()

X = df.iloc[:,0].values.reshape(-1,1)
Y = df.iloc[:,1].values.reshape(-1,1)

X_design = np.hstack([np.ones((X.shape[0],1)), X])

beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

# Prediction
pred = np.array([1,12]) @ beta
print("Predicted Salary:", pred[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('hiring.csv')

# ---------- Data Cleaning ----------

# Convert word experience to numbers
word_to_num = {
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11,
    'twelve':12
}

df['experience'] = df['experience'].map(word_to_num)

# Fill missing experience with 0
df['experience'] = df['experience'].fillna(0)

# Fill missing test scores with median
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

# ---------- Features ----------
X = df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']]

y = df['salary($)']

# ---------- Train Model ----------
model = LinearRegression()
model.fit(X, y)

# ---------- Predictions ----------
pred1 = model.predict([[2,9,6]])
pred2 = model.predict([[12,10,10]])

print("Salary Prediction 1:", pred1[0])
print("Salary Prediction 2:", pred2[0])
```

```python

import pandas as pd
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv('1000_Companies.csv')

# ----- Data preprocessing -----
# One-hot encode state column
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Features and target
X = df.drop('Profit', axis=1)
y = df['Profit']

# ----- Train model -----
model = LinearRegression()
model.fit(X, y)

# ----- Prediction -----
# IMPORTANT: column order must match X columns
input_data = [[91694.48, 515841.3, 11931.24, 0, 1]]
# (Example encoding: Florida=0, New York=1 etc — adjust based on your columns)

pred_profit = model.predict(input_data)

print("Predicted Profit:", pred_profit[0])
```
